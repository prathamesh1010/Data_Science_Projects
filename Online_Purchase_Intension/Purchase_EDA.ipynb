## Distribution of Target Variable

```python
target_count = data['Target'].value_counts()
plt.figure(figsize=(8,4))
sns.countplot(x='Target', data=data)
plt.title('Distribution of Target Variable')
plt.show()
```

## Histogram for 'BounceRates' and 'ExitRates'

```python
numerical_cols = ['BounceRates', 'ExitRates']
#Replace with actual column names
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(data[col])
    plt.title(f'Historam of {col}')
    plt.show()
```

## Correlation Matrix for BounceRate and ExitRate

```python
correlation_matrix = data[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Correlation Matrix')
plt.show()
```

## Correlation Matrix Heatmap Including All Parameters

```python
#Making All Categorical values into numerical values for Correlation Heatmap 
data['Month'] = data['Month'].map({
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'June': 6, 
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
})
data['VisitorType'] = data['VisitorType'].map({
    'New_Visitor': 0, 'Returning_Visitor': 1, 'Other': 2
})
data['Target'] = data['Target'].astype(int)
plt.figure(figsize=(12, 10))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Heatmap Including All Parameters")
plt.show()
```

## ## Important parameters to predict the target variable
#### The heatmap indicates a strong correlation between Target and features such as Month, ProductRelated, ProductRelated_Duration, VisitorType and Pagevalue. In the model development process, we have selected Pagevalue as a predictor for the Target variable.
